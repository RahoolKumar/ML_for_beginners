```python
import pandas as pd 

nba = pd.read_csv("csv/nba.csv")
nba.head()
```

```python
nba["Salary"].add(50000)
nba["Age"].sub(1)
nba["Weight"].mul(0.5)
nba["Weight in KG"] = nba["Weight"].mul(0.459)
nba.head()
```

```python
nba.tail()
```

```python
# to remove row with col na
nba.dropna().head(7)
```

```python
nba.dropna(how="any").head(7)
# this works same as nba.dropna()
```

```python
# to remove row whose all columns values are null
nba.dropna(how="all").head(7)
```

```python
nba.dropna(how="all", inplace=True)
nba.tail()
```

```python
# to remove NaN from every column
nba.dropna(axis=1).head()
```

```python
# to remove Nan value from a specific column
nba.dropna(subset=["Salary"]).head()
```

```python
# to fill a specific value in replacement of na 
nba.fillna(0).head()
```

```python
nba["Salary"]
```

```python
nba["Salary"].fillna(0, inplace=True)
```

```python
nba.head()
```

```python
nba["College"].fillna("No College", inplace=True)
nba.head(8)
```

```python
nba.sort_values("Name").head(7)
```

```python
nba.sort_values("Age", ascending=False).head(7)
```

```python
nba1 = pd.read_csv("csv/nba.csv")
nba1.sort_values("Salary", na_position="first").head(7)
```
